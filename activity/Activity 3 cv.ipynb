```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Read image (PUT your exact uploaded filename)
img = cv2.imread("Screenshot 2026-02-10 143249 (2).png")

# Check image
if img is None:
    print("❌ Image not found.")
else:
    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Apply Gaussian blur (smooth image)
    blur = cv2.GaussianBlur(gray, (5,5), 0)

    # Threshold → creates REGION mask
    _, thresh = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)


    # Create colored region output
    region = np.zeros_like(img)
    region[thresh == 255] = (0, 165, 255)   # ORANGE filled region

    # Show outputs
    print("Original Image")
    cv2_imshow(img)

    print("Region Mask")
    cv2_imshow(thresh)

    print("Region-Based Segmentation (Orange Filled)")
    cv2_imshow(region)
```
